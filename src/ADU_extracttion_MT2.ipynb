### Extract Relation

```python
# Sample input structure
# Assuming `myList` is a list of elements, where each element contains nested 'edge' elements with attributes 'src', 'tgt', and 'type'
import xml.etree.ElementTree as ET

# Example XML data for demonstration
data = '''
<root>
    <item>
        <edge src="a1" tgt="a2" type="sub"/>
        <edge src="a2" tgt="a3" type="reb"/>
        <edge src="a3" tgt="a4" type="other"/>
    </item>
    <item>
        <edge src="a4" tgt="a5" type="sub"/>
        <edge src="a5" tgt="a6" type="reb"/>
        <edge src="a6" tgt="a7" type="seg"/>
    </item>
</root>
'''

# Parse the XML data
root = ET.fromstring(data)
myList = root.findall('.//item')

# Initialize the list Z to store edges of type 'sub' or 'reb'
Z = []

# Iterate through each element in myList
for k in myList:
    # Extract edges with type 'sub' or 'reb'
    for e in k.findall('.//edge'):
        if e.attrib['type'] in ['sub', 'reb']:
            # Append the src, tgt, and type to list Z
            Z.append([e.attrib['src'], e.attrib['tgt'], e.attrib['type']])

# Now Z contains all edges of type 'sub' or 'reb' with their src, tgt, and type
print("Extracted edges:", Z)
```

```python
import os
import xml.etree.ElementTree as ET

def process_xml_file(file_path):
    try:
        tree = ET.parse(file_path)
        root = tree.getroot()

        relations = []
        for edge in root.findall('edge'):
            src = edge.get('src')
            trg = edge.get('trg')
            type_rel = edge.get('type')

            if type_rel in ['reb', 'und', 'add', 'sup']:
                relations.append([src, trg, type_rel])

        return relations
    except ET.ParseError as e:
        print(f"Error parsing {file_path}: {e}")
        return []

def process_all_xml_files(directory):
    all_relations = []
    for filename in os.listdir(directory):
        if filename.endswith('.xml'):
            file_path = os.path.join(directory, filename)
            relations = process_xml_file(file_path)
            all_relations.append(relations)
    
    return all_relations

directory_path = path  # Replace with your actual directory path
all_relations = process_all_xml_files(directory_path)
print(all_relations)
```

```python
nlp = spacy.load("en_core_web_sm")
def mapTokenPos(l):
    tokenlist=[]
    for x in l:
        doc=nlp(x)
        a=[(token.text,token.tag_) for token in doc]
        tokenlist.append(a)
    return tokenlist
```

```python
df["IOB"]=df["EDU"].apply(mapTokenPos)
```

```python
def mapLabel(x):
    n=[]
    for l in x:
        if l==0:
            n.append("P")
        elif l==1:
            n.append("C")
    return(n)
```

```python
df["Label"]=df["Label"].apply(mapLabel)
```

```python
def mapIOB(a,b):
    tlist=[]
    for lst,label in zip(a,b):
        c=(lst,label)
        tlist.append(c)
    return(tlist)

```

```python
df["IOB"]=df.apply(lambda  x :mapIOB(x["IOB"], x["Label"]),axis=1)
```

```python
def mapPC(d):
    new_d =  [ [t+('BI'[i>0]+'-'+lb,) for i,t in enumerate(T)] for T,lb in d ]
    return(new_d)
```

```python
df["IOB"]=df["IOB"].apply(mapPC)
```

```python
def flatIOB(l):
    f = [item for t in l for item in t]
    return(f)
```
