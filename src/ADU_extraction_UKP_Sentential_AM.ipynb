```python

data = pd.read_json(path)

parsed_data = {
    "topic_id": [],
    "EDU": [],
    "adu_type": [],
    "targetsRelation": [],
    "Compatible": []
}

for _, row in data.iterrows():
    debate_topic = row["topicTarget"]
    claims = [claim["claimOriginalText"] for claim in row["claims"]]
    stances = [claim["stance"] for claim in row["claims"]]
    
    targets_relation = [claim.get("targetsRelation", None) for claim in row["claims"]]
    compatible = [claim.get("Compatible", None) for claim in row["claims"]]
    
    parsed_data["topic_id"].append(debate_topic)
    parsed_data["EDU"].append(claims)
    parsed_data["adu_type"].append(stances)
    parsed_data["targetsRelation"].append(targets_relation)
    parsed_data["Compatible"].append(compatible)

df = pd.DataFrame(parsed_data)
df.head()
```

```python
df.to_pickle('..\data\df_IBM.pkl')
df.to_json('..\data\df_IBM.json')
```
